# Find largest & smallest element


```python
n = int(input("Enter list size:"))
num = []
for i in range(n):
    value = int(input("Enter value:"))
    num.append(value)
max = num[0];
min = num[0];
for i in range(n):
    if(num[i] > max):
        max = num[i]
    else:
        min = num[i]
print("Larjest Number:",max)
print("Smallest Number:",min)    
```

# Remove duplicate elements

```python
n = int(input("Enter List Size:"))
num = []
for i in range(n):
    value = int(input("Enter Value:"))
    num.append(value)
unique = []
for i in num:
    if num not in unique:
        unique.append(i)
print(unique)        
```

# Sort list without sort()

```python
n = int(input("Enter List size: "))
num = []

for i in range(n):
    value = int(input("Enter a value: "))
    num.append(value)

for i in range(n):
    for j in range(0, n - i - 1):
        if num[j] > num[j + 1]:
            temp = num[j]
            num[j] = num[j + 1]
            num[j + 1] = temp

print(num)
```

# Find second largest element

```python
n = int(input("Enter List size: "))
num = []

for i in range(n):
    value = int(input("Enter a value: "))
    num.append(value)
max = num[0];
smax = -1;
for i in range(n):
    if(num[i] > max):
        smax = max;
        max = num[i];
    elif(num[i] < max and num[i] > smax):
        smax = num[i];
print("Second Largest Element:",smax)
```

# Reverse a list

```python
n = int(input("Enter List size: "))
num = []

for i in range(n):
    value = int(input("Enter a value: "))
    num.append(value)
rev = []    
for i in range(n-1,-1,-1):
    rev.append(num[i])
print(rev)    
```

# Count even and odd numbers

```python
n = int(input("Enter List size: "))
num = []

for i in range(n):
    value = int(input("Enter a value: "))
    num.append(value)
even = 0
odd = 0
for i in num:
    if(i % 2 == 0):
        even = even+1;
    else:
        odd = odd+1;
print("counted Even Number:",even)
print("counted Odd Number:",odd)        
```

# Find common elements in two lists

```python
l1 = [1,2,3]
l2 = [2,3,6]
res = []
for i in l1:
    if i in l2:
        res.append(i);
print(res)        
```

# Rotate list by K positions

```python
n = int(input("Enter List size: "))
num = []

for i in range(n):
    value = int(input("Enter a value: "))
    num.append(value)
k = int(input("Enter K Position:"))    
k = k % n;
rotated = num[k:]+num[:k]
print(rotated)
```
