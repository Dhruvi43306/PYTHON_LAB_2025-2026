### 13) WAP to calculate area and perimeter of a circle.

```python
import math

r = float(input("Enter a Redius:"))
area = math.pi * r * r
perimeter = 2 * math.pi * r
print(f"Area of Circle :{area:.2f}")
print(f"Perimeter of Circle :{perimeter:.2f}")
```

### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
print("1.Degree to Fahrenheit")
print("2.Fahrenheit to Degre")
choise = int(input("Enter a Choise(1 or 2):"))
if choise == 1:
    c = float(input("Enter Temperature in Degree."))
    f = (c* 9/5)+32
    print(f"Degree to Fahrenheit:{f : .2f}°F")
elif choise == 2:
    f = float(input("Enter Temperature in Fahrenheit."))
    c = (f-32)*5/9;
    print(f"Fahrenheit to Degree:{c : .2f}°C")
else:
    print("Invalid Input!!")
```
