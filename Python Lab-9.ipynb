## Practical 1: Student Marks Analyzer

### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks and store result in a dictionary.

### Sample Input
```python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
```

### Sample Output
```
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```

````python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
res = {};
for i, j, k in students:
        k = set(k)
        avg = round(sum(k)/len(k),2)
        res[i] = (j,avg)
print(res)   
````

## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
res = {}
sum1=0
total = 0;
for k,v in inventory.items():
    if v[1] != 0:
        res[k] = v
        sum1 = v[0]*v[1]
        total += sum1;

print(res)
print("Total value:",total)
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
res = []
freq = {};
for i in emails:
    s = i.split("@")
    if s[1] not in freq:
        res.append(s);
        freq[s[1]] = 1
    else:
        freq[s[1]] += 1            
print(freq)        
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
carts = dict(cart)
res = {}
sum1 = 0
for k1,v1 in carts.items():
    for k2,v2 in prices.items():
        if k1 == k2:
            total = v1*v2;
            sum1 += total
            res[k1] = total
print(res)
print("Total Bill:",sum1)
````

## Practical 5: Character Position Index

### Problem Definition
Store positions of each character in a string.

### Sample Input
```python
s = "banana"
```

### Sample Output
```
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}
```

````python
s = "banana"
res = {}
for i,char in enumerate(s):
      if char not in res:
          res[char] = []
      res[char].append(i)
print(res)    
````

## Practical 6: Survey Data Aggregation

### Problem Definition
Count responses city-wise and choice-wise.

### Sample Input
```python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
```

### Sample Output
```
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}
```

````python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
res = {}
for i,j,k in responses:
    print(i,j,k)
    if j not in res:
        res[j] = {}
    if k not in res:
        res[j][k] = 1
print(res)

       
````
