### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s = input("Enter String:");
upper = 0;
lower = 0;
digit = 0;
special = 0;
for ch in s:
    if(ch.isupper()):
        upper+=1;
       
    elif(ch.islower()):
        lower+=1;
        
    elif(ch.isdigit()):
        digit+=1;
    else:
        special+=1
        
print("Uppercase:",upper);
print("wercase:",lower)
print("Digit:",digit)
print("Special:",special)
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input("Enter String:");
count = 1;
res = ""
for i in range(1,len(s)):
    if s[i] == s[i-1]:
        count += 1;
    else:
        res = res + s[i-1] + str(count)
        count = 1
res += s[-1] + str(count);
print(res)
```

### 15) Custom Split
- `apple,banana,grape`
- `10|20|30`
- `A-B-C-D`

```python
s = input("Enter String:")
delimeter = input("Enter Delimeter:")
res = []
temp = ""
for ch in s:
    if ch == delimeter:
        res.append(temp)
        temp = ""
    else:
        temp+=ch
res.append(temp)        
print(res)        
```
