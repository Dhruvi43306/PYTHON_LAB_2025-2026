### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter a Number:"))
sum = 0;
original_no = n;
while(n > 0):
    rem = n % 10;
    sum = sum*10 + rem;
    n = n // 10;
if(sum == original_no):
     print("Palingdrome.")
else:
    print("Not Palingdrome.")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = int(input("Enter a Number:"))
sum = 0;
original_no = n;
a = len(str(n))
while(n > 0):
    rem = n %10;
    n = n // 10;
    sum = sum + (rem**a)
if(sum == original_no):
    print("Armstrong")
else:
    print("Not Armstrong");
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input("Enter a First Number:"))
for i in range(1,n+1):
    sum = 0;
    for j in range(1,i):
        if(i % j == 0):
            sum += j;
    if(sum == i):
        print(f"{i} is Perfect Number.");
   
    
```
