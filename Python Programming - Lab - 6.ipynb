### 01) WAP to find sum of all the elements in a List.

```python
n = int(input("Enter List size:"));
num = [];
sum = 0;
for i in range(n):
    value = int(input("Enter List value:"));
    num.append(value)
print(num)
for i in range(n):
    sum = sum + num[i];
print("Sum:",sum);    
```

### 02) WAP to find largest element in a List.

```python
n = int(input("Enter List size:"));
num = [];
for i in range(n):
    value = int(input("Enter List value:"));
    num.append(value)
max = num[0]
for i in range(n):
    if(num[i] > max):
        max = num[i];
print("Largest Element:",max)        
```

### 04) WAP to interchange first and last elements in a list.

```python
n = int(input("Enter List size:"));
num = [];
for i in range(n):
    value = int(input("Enter List value:"));
    num.append(value)
    temp = num[0];
    num[0] = num[-1]
    num[-1] = temp;
print(num)
```

### 09) WAP to count unique items in a list.

```python
n = int(input("Enter List size:"));
num = [];
count = 0;
for i in range(n):
    value = int(input("Enter List value:"));
    num.append(value)
print("Original List:",num) 
for i in num:
    x = 0
    for j in num:
        if(i == j):
            x = x+1
    if(x == 1):
        count = count+1;
print(count);        
```

### 13) WAP to find second largest number in a list.

```python
n = int(input("Enter List size: "))
num = []

for i in range(n):
    value = int(input("Enter a value: "))
    num.append(value)
max = num[0];
smax = -1;
for i in range(n):
    if(num[i] > max):
        smax = max;
        max = num[i];
    elif(num[i] < max and num[i] > smax):
        smax = num[i];
print("Second Largest Element:",smax)    
```

### 14) WAP to extract elements with frequency greater than K.

```python
n = int(input("Enter List size: "))
num = []
res = []
for i in range(n):
    value = int(input("Enter a value: "))
    num.append(value)
fre = int(input("Enter a Frequency:"))    
for i in num:
    x = 0
    for j in num:
        if(i == j):
            x = x+1;
    if(x > fre) and i not in res:
        res.append(i)
print(res)        
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
n = int(input("Enter List size: "))
num = []

for i in range(n):
    value = input("Enter a value: ")
    num.append(value)
res = [i for i in num if(i.lower().startswith("b"))]
print(res)
```
